#Importing libraries & creating Lists for txt files #



```python
import numpy as np
import os
import glob
import nltk
import re
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import string
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
from sklearn.feature_extraction.text import CountVectorizer
import scipy.sparse as sp


pos_path = "pos"
neg_path = "neg"

pos_files = glob.glob(os.path.join(pos_path, "*.txt"))
neg_files = glob.glob(os.path.join(neg_path, "*.txt"))
```

```python
pos_texts = []
for file_path in pos_files:
    with open(file_path, "r") as f:
        text = f.read()
        pos_texts.append(text)

neg_texts = []
for file_path in neg_files:
    with open(file_path, "r") as f:
        text = f.read()
        neg_texts.append(text)
```

```python
print(pos_texts[0]) #each index in the list contains a text file.
```

#Tokenization

```python
pos_tokens = []
for text in pos_texts:
    tokens = nltk.word_tokenize(text)
    pos_tokens.append(tokens)

neg_tokens = []
for text in neg_texts:
    tokens = nltk.word_tokenize(text)
    neg_tokens.append(tokens)
```

```python
print(pos_tokens[0])
print(neg_tokens[0])
```

#Pre-Processing & Cleaning


**The following Code:**
*   Remove stopwords
*   Remove punctuation marks
*  apply Porter stemmer









```python
stop_words = set(stopwords.words('english'))


# Define a regular expression to match words
regex =  re.compile('\w+')

pos_cleaned = []
for tokens in pos_tokens:# access the positive files
    cleaned_tokens = []  # the file after being cleaned
    for token in tokens:#access te positive text
       if token.lower() not in stop_words and token.lower() not in string.punctuation: #remove stop words & punctuation marks
            token = token.replace('``', '')
            token = token.replace('--', '')
            cleaned_token = regex.findall(token.lower())
            cleaned_token = ' '.join(cleaned_token)
            stemmedword = porter.stem(token.lower())
            if stemmedword != '':
              cleaned_tokens.append(stemmedword)
    pos_cleaned.append(cleaned_tokens)#adding the cleaned file to list of positive cleaned files

neg_cleaned = []
for tokens in neg_tokens:
    cleaned_tokens = []
    for token in tokens:
        if token.lower() not in stop_words and token.lower() not in string.punctuation:
            token = token.replace('``', '')
            token = token.replace('--', '')
            cleaned_token = regex.findall(token.lower())
            cleaned_token = ' '.join(cleaned_token)
            stemmedword = porter.stem(token.lower())
            if stemmedword != '':
              cleaned_tokens.append(stemmedword)
    neg_cleaned.append(cleaned_tokens)
```

```python
print(pos_cleaned[0])
print(neg_cleaned[0])
```

#Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Combine positive and negative comments
comments = [' '.join(tokens) for tokens in pos_cleaned + neg_cleaned]
# Create labels for positive and negative comments
labels = ['pos'] * len(pos_cleaned) + ['neg'] * len(neg_cleaned)

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the comments
vectorized_comments = tfidf_vectorizer.fit_transform(comments)
x=vectorized_comments
# Add the 'type' column
type_column = np.array(labels).reshape(-1, 1)
vectorized_comments_with_type = np.hstack((vectorized_comments.toarray(), type_column))

```

#Train-Test-Split

```python
y = vectorized_comments_with_type[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train, y_test = train_test_split(x,y, random_state=104, test_size=0.33, shuffle=True)
```

```python

from sklearn import svm
C = 0.1
# SVM regularization parameter
svc = svm.SVC(kernel='linear', C=C).fit(X_train, y_train)
lin_svc = svm.LinearSVC(C=C).fit(X_train, y_train)
rbf_svc = svm.SVC(kernel='rbf', gamma=0.8, C=C).fit(X_train, y_train)
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X_train, y_train)

# title for the plots
titles = ['SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']


for i, clf in enumerate((svc, lin_svc, rbf_svc, poly_svc)):
    predictions = clf.predict(X_train)
    predictions = np.asarray(predictions).reshape(-1,1)
    accuracy = accuracy_score(y_train,predictions)*100
    print(titles[i], 'train accuracy: ', accuracy)

    predictions = clf.predict(X_test)
    predictions = np.asarray(predictions).reshape(-1,1)
    accuracy = accuracy_score(y_test,predictions)*100
    print(titles[i], 'test accuracy: ', accuracy, '\n')

    print(classification_report(y_test , predictions))
    print(confusion_matrix(y_test, predictions))
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier,OneVsOneClassifier
import numpy as np

lr_ovo = OneVsOneClassifier(LogisticRegression()).fit(X_train, y_train)
lr_ovr = OneVsRestClassifier(LogisticRegression()).fit(X_train, y_train)


# model accuracy for Logistic Regression model
accuracy = lr_ovr.score(X_test, y_test) * 100
print('OneVsRest Logistic Regression accuracy: ' + str(accuracy))
print(classification_report(y_test , lr_ovr.predict(X_test)))
print(confusion_matrix(y_test, lr_ovr.predict(X_test)))
accuracy = lr_ovo.score(X_test, y_test) * 100
print('OneVsOne Logistic Regression accuracy: ' + str(accuracy))
print(classification_report(y_test , lr_ovo.predict(X_test)))
print(confusion_matrix(y_test, lr_ovo.predict(X_test)))

```
